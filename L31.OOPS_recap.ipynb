```python
class emp:
    def __init__(self,name,email):
        self.name = name
        self.email = email
    
    def displayemp (self):
        print(f"the name of the employee is {self.name}")
        print(f"the email of the employee is {self.email}")

abhi = emp('mahi','mai@gmail.com')
nancy = emp('nanc','nany@gmail.com')
```

```python
abhi.displayemp()
nancy.displayemp()
```
